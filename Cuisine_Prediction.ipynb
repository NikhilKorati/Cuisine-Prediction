# Exploratory Data Analysis

```python
df.country = df.country.str.upper()
```

```python
df.replace(to_replace='CHINA', value='CHINESE', inplace=True)
df.replace(to_replace='FRANCE', value='FRENCH', inplace=True)
df.replace(to_replace='GERMANY', value='GERMAN', inplace=True)
df.replace(to_replace='INDIA', value='INDIAN', inplace=True)
df.replace(to_replace='ITALY', value='ITALIAN', inplace=True)
df.replace(to_replace='JAPAN', value='JAPANESE', inplace=True)
df.replace(to_replace='KOREA', value='KOREAN', inplace=True)
df.replace(to_replace='MEXICO', value='MEXICAN', inplace=True)
df.replace(to_replace='THAILAND', value='THAI', inplace=True)
df.replace(to_replace='SCANDINAVIA', value='SCANDINAVIAN', inplace=True)
```

```python
value_count = df.stack().value_counts()
removed = value_count[value_count < 75].index
df = df.loc[~df.country.isin(removed)]
df.shape
```

```python
df[(df.rice == 'Yes') & (df.turmeric ==  'Yes') & (df.cumin == 'Yes')]
```

```python
(df == 'Yes').sum().sort_values(ascending = False)
```

```python
by_country_df = (df == 'Yes').groupby(df.country).mean()
by_country_df
```

```python
by_country_df.loc['INDIAN'].sort_values(ascending = False)
```

```python
region = by_country_df.index.get_values()

for name_of_region in region:
    print(name_of_region)
    print(by_country_df.loc[name_of_region].sort_values(ascending = False).ix[0:5])
    print('\n')
```

# K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
no_of_clusters = 5
```

```python
y_pred = KMeans(n_clusters= no_of_clusters, random_state=0).fit_predict(by_country_df)
```

```python
for i in range(0, no_of_clusters):
    print('Cluster: '+str(i))
    print('------------------------------------------------------------------------------------------')
    print(list(by_country_df[y_pred == i].index))
    print('------------------------------------------------------------------------------------------')
    print('\n')
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
import matplotlib.image as mpimg

def visualize_tree(decisionTreeModel, X, row_values, filename):
    dot_data = StringIO()
    out = tree.export_graphviz(decisionTreeModel, feature_names=X, out_file=dot_data, class_names=row_values, 
                              filled = True, rounded = True, node_ids = True, proportion = True, 
                              special_characters = True, impurity = True, label = 'all', leaves_parallel = False)
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    graph.write_png(filename)
    img = mpimg.imread(filename)
    fig = plt.figure(figsize=(55,16), dpi = 50, facecolor='w', edgecolor='k')
    plt.imshow(img)
```

```python
visualize_tree(decisionTree, list(df.ix[:,1:].columns.values), np.unique(y), 'abc.png')
```
